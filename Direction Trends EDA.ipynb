```python
df_centrality_full = pd.read_csv('Github_degree_betweenness_closeness.csv')
df_centrality_cross = pd.read_csv('Centrality_github_data_cross_edge.csv')

df_compare = df_centrality_cross.merge(df_centrality_full, how = 'left', right_on = 'id',\
                                      left_on = 'id')
df_compare.head()

#x is cross edge
```

```python
df_compare['Centrality_diff'] = df_compare['Centrality_std_x'] - df_compare['Centrality_std_y']
df_compare.head()
```

```python
from sklearn import preprocessing

df_compare['Centrality_diff_std'] = preprocessing.scale(df_compare['Centrality_diff'])
df_compare.head()
```

```python
df_compare = df_compare.sort_values(by='Centrality_diff_std', ascending=False)
df_compare
#larger scores mean that cross edge has a higher centrality - these are the most collaborative people
#relative to their group
```

```python
df_compare[df_compare['ml_target_x'] == 1]
```

```python
#import matplotlib
#df_compare['Centrality_std_y'].hist()
df_filtered =df_compare[df_compare['Centrality_std_y'] >= 0.1]
df_filtered = df_filtered[df_filtered['Degree_betweenness_closeness'] >=0.1]# and df_compare['Centrality'] >= 0.1]
```

```python
df_filtered['Centrality_perc_diff'] = (df_filtered['Centrality_std_x'] \
                                       - df_filtered['Centrality_std_y']) /df_filtered['Centrality_std_y']
```

```python
df_filtered.sort_values(by='Centrality_perc_diff')
```

```python
df_new = pd.read_csv('combined_github_data.csv')
df_new.head()
```

```python
df_new['Cross edge'] = df_new['ml_target'] != df_new['ml_target_2']
df_new.head()
```

```python
import numpy as np
#df_new['name_count']= 
def count(x):
    return df_new.groupby('name')['Cross edge'].agg(sum).sort_values()[x]

#df_new = df_new.set_index('name')
df_new.head()
for name in df_new.index.values:
    if count(name) <= 5:
        df_new.drop(name, axis=0)
        
df_new.head()
```

```python
name_means = pd.Series(df_new.groupby('name')['Cross edge'].agg(np.mean).sort_values())
name_sums = pd.Series(df_new.groupby('name')['Cross edge'].agg(sum))

len(name_means)
mask = name_sums > 75

name_means = name_means[mask]
name_means
```
